**Image to Pencil Sketch with Python**

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/photo.jpg.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
img.min()
```

```python
inv = 242-img
```

```python
plt.imshow(inv, cmap = 'gray')
```

```python
blur = cv2.GaussianBlur(inv, ksize=(21,21), sigmaX = 0, sigmaY = 0)
```

```python
plt.imshow(blur, cmap = 'gray')
```

```python
dodge = lambda image, mask: cv2.divide(image, 242-mask, scale=256 )
```

```python
blended = dodge(img, blur)
```

```python
plt.imshow(blended, cmap = 'gray')
```

```python
cv2.imwrite('imgsketch.jpg', blended)
```
